1. Import your libraries and create your dataframe.


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

url= '/content/Nutrition_Facts_for_McDonald_s_Menu_1603_69.csv'
Nutrition_Facts = pd.read_csv(url)
Nutrition_Facts.head()
```

5. What item has the most calories? What has the least?

```python
High_calorie = Nutrition_Facts.groupby("Item")["Calories"].sum().sort_values(ascending=True).head(1)
Low_calorie = Nutrition_Facts.groupby("Item")["Calories"].sum().sort_values(ascending=False).head(1)
High_calorie, Low_calorie
```

6. How many items have more than 10 grams of protein?

```python
items_more_than_10_grams = Nutrition_Facts[Nutrition_Facts['Protein'] > 10].shape[0]
items_more_than_10_grams
```

7. Create a bar graph of the top 5 highest calorie items.

```python
top_5_highest_calorie_items = Nutrition_Facts.groupby('Item')['Calories'].sum().sort_values(ascending=False).head()
top_5_highest_calorie_items.plot(kind='bar', title='Top 5 highest calorie items')
plt.xlabel('Item')
plt.ylabel('Calories')
plt.show()
```

9. Create a pie chart showing the number of items in each meal category

```python
items_in_meal = Nutrition_Facts.groupby('Category')['Category'].count().sort_values(ascending=False)
items_in_meal.plot(kind='pie', title='Number of items in each meal category')
plt.show()
```

10. Come up with your own visualization! It can be about anything in this dataset.

```python
top_5_highest_sodium_items = Nutrition_Facts.groupby('Item')['Sodium'].sum().sort_values(ascending=False).head()
top_5_highest_sodium_items.plot(kind='bar', title='Top 5 highest sodium items')
plt.xlabel('Item')
plt.ylabel('Sodium')
plt.show()
```
